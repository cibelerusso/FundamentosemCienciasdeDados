## O método científico: Métodos, Técnicas e Algoritmos

Antes de começarmos a programar ou aplicar modelos em ciência de dados, é importante compreender alguns conceitos fundamentais que aparecem com frequência na literatura e na prática: **métodos**, **técnicas** e **algoritmos**.

Embora esses termos sejam frequentemente utilizados como sinônimos no dia a dia, eles possuem significados distintos.


Existe uma distinção clara entre esses termos, embora eles estejam intimamente relacionados no contexto da análise de dados e programação:

-   **Método ou Técnica:** É definido como um **procedimento sistemático** que permite alcançar um objetivo pretendido. O método descreve, de forma geral, como realizar uma determinada tarefa. Por exemplo, o cálculo da média de idades de um grupo é um método.

-   **Algoritmo:** É um conjunto de **instruções passo a passo**, auto-contido e facilmente compreensível por humanos, que permite a **implementação** de um determinado método. Os algoritmos são projetados para serem facilmente traduzidos para qualquer linguagem de programação.

### Relação entre eles
A principal diferença é que o método mostra o "que" fazer e o caminho sistemático, enquanto o **algoritmo fornece as instruções detalhadas** para que esse método seja executado por um computador.

**Exemplo Prático:**

-  **Método:** Obter a média aritmética de uma lista de números.

- Algoritmo

    1. **Iniciar** uma variável chamada `soma` com valor igual a **0**.  
    2. **Calcular** o número de elementos da lista e armazenar em uma variável `n`.  
    3. **Percorrer** todos os elementos da lista.  
        3.1. **Adicionar** cada elemento da lista à variável `soma`.  
    4. **Dividir** o valor total da `soma` pelo número de elementos `n`.  
    5. **Retornar** o resultado da divisão como a média da lista.

Em ciência de dados, técnicas como o **k-means** são aplicadas através de algoritmos específicos que seguem passos repetitivos (como associar instâncias a centroides e recalcular esses centroides) até que o objetivo do método seja atingido.

Além disso, as fontes destacam que a **lógica de programação** (essencial para criar algoritmos) é considerada mais importante do que qualquer linguagem de programação específica, pois uma vez aprendida, ela permite ao programador resolver problemas em diversas linguagens.

## Método e Algoritmo: Exemplo da Pipoca 🍿

Para entender a diferença entre **método** e **algoritmo**, vamos usar um exemplo simples do cotidiano: **estourar pipoca**.

### Método

Um **método** é um **procedimento geral** utilizado para alcançar um objetivo.  
Ele descreve **a ideia ou estratégia**, mas não necessariamente todos os detalhes.

**Exemplo: método para estourar pipoca**

> Aquecer os grãos de milho até que atinjam a temperatura necessária para estourar, utilizando uma fonte de calor adequada e evitando que queimem.

Observe que o método explica **o que deve ser feito**, mas não especifica exatamente **cada passo**.



### Algoritmo

Um **algoritmo** é um **conjunto finito de instruções passo a passo** que descreve exatamente **como executar uma tarefa**.

**Exemplo: algoritmo para estourar pipoca**

1. Pegue uma panela com tampa.
2. Coloque a panela no fogão.
3. Adicione **2 colheres de sopa de óleo**.
4. Ligue o fogo em temperatura média.
5. Adicione **1/2 xícara de milho de pipoca**.
6. Tampe a panela.
7. Aguarde os primeiros estouros.
8. Mexa a panela levemente de tempos em tempos para evitar que a pipoca queime.
9. Quando o intervalo entre os estouros for maior que **2 ou 3 segundos**, desligue o fogo.
10. Abra a tampa com cuidado.
11. Coloque a pipoca em um recipiente.
12. Adicione sal ou tempero a gosto.



### Resumindo

| Conceito | Descrição |
|--------|--------|
| **Método** | Estratégia geral para resolver um problema |
| **Algoritmo** | Passo a passo detalhado para executar a solução |

No exemplo:

- **Método:** aquecer o milho até ele estourar.  
- **Algoritmo:** sequência detalhada de passos para produzir a pipoca.

## Pensamento Científico na Ciência de Dados


Com base nas fontes fornecidas, o **método científico** aplicado à análise de dados é fundamentado na **estatística**, especificamente na **aprendizagem indutiva**, que busca generalizar conhecimento a partir de amostras. A **Analytics** é definida como a ciência que analisa dados brutos para extrair conhecimento útil (padrões).

Abaixo, os principais componentes do pensamento científico na ciência de dados:

### 1. Indução e Dedução
*   **Indução (Inferência Estatística):** É o processo de generalizar o conhecimento obtido de uma amostra para toda a população. Como as populações costumam ser muito grandes ou proibitivamente caras para serem totalmente pesquisadas, a ciência utiliza a amostragem para estimar valores populacionais.
*   **Dedução:** É o inverso da indução; ela particulariza da população para a amostra. As probabilidades, por exemplo, tratam de problemas dedutivos ao tentar prever a natureza de uma amostra sabendo-se as características da população.

### 2. Metodologias Estruturadas
O método científico em dados não é aleatório; ele segue frameworks rigorosos para garantir a validade do conhecimento descoberto:

*   **Processo KDD (Knowledge Discovery from Data):** Uma metodologia acadêmica que propõe nove etapas sequenciais, embora permita retroceder passos para refinar o processo. As etapas incluem aprender o domínio do problema, limpar e pré-processar os dados, escolher algoritmos de mineração e, finalmente, interpretar os resultados para uso prático.
*   **CRISP-DM:** Uma metodologia voltada para a indústria que organiza o projeto em seis fases cíclicas:

- compreensão do negócio,
- compreensão dos dados,
- preparação dos dados,
- modelagem,
- avaliação e
- implantação.

### 3. Métodos e Algoritmos
A ciência de dados distingue o conceito abstrato da execução técnica:
*   **Método (ou Técnica):** É um procedimento sistemático para atingir um objetivo pretendido.
*   **Algoritmo:** É um conjunto de instruções passo a passo, autocontido e compreensível, que permite a implementação de um método. Um algoritmo descreve, por exemplo, como calcular matematicamente a média de uma amostra ou como induzir um modelo de previsão.

### 4. Modelagem e Generalização
O objetivo final do método científico em dados é a **indução de modelos**. Um modelo é uma generalização que pode ser usada como protótipo para gerar previsões sobre novos objetos desconhecidos. A ciência busca minimizar erros de previsão futuros através do treinamento e validação desses modelos em conjuntos de dados distintos.

Em resumo, a abordagem científica na programação e análise de dados exige **lógica**, **paciência** e **atenção aos detalhes**, pois a lógica de programação é considerada mais importante do que qualquer linguagem específica para expressar soluções científicas.

## Aplicando o método científico na análise de dados

Uma forma comum de aplicar o método científico em ciência de dados é usar metodologias estruturadas.  
Uma das mais conhecidas é a **CRISP-DM (Cross Industry Standard Process for Data Mining)**.

Para ilustrar esse processo, vamos usar um exemplo clássico: o **conjunto de dados de câncer de mama de Wisconsin**.



### 1. Compreensão do problema

O primeiro passo é definir o problema.

Imagine que um hospital deseja um **sistema de apoio ao diagnóstico de câncer de mama**.

A pergunta central é:

> Existem padrões nas características das células que permitam diferenciar tumores benignos de malignos?

Essa etapa define o objetivo da análise.



### 2. Compreensão dos dados

Em seguida, analisamos os dados disponíveis.

No dataset de Wisconsin, os dados vêm de **imagens digitalizadas de biópsias por aspiração com agulha fina**.

Cada observação descreve características das células, como:

- espessura do aglomerado celular  
- uniformidade do tamanho das células  
- uniformidade da forma das células  
- presença de núcleos nus  

Também verificamos a qualidade dos dados.  
Por exemplo, o atributo **"núcleos nus" possui valores ausentes**.



### 3. Preparação dos dados

Antes da modelagem, os dados precisam ser preparados.

As principais etapas incluem:

- **Limpeza:** remover registros com valores ausentes  
- **Seleção:** excluir variáveis que não ajudam na análise (ex.: ID da amostra)  
- **Escala:** verificar se as variáveis precisam ser normalizadas  

Neste dataset, as variáveis já estão na escala **1 a 10**, então não foi necessário normalizar.



### 4. Modelagem

Agora aplicamos técnicas para identificar padrões.

No exemplo, utiliza-se o **k-means**, um algoritmo de **agrupamento (clustering)**.

O pesquisador testa diferentes valores de \(k\), que representa o número de grupos.

O objetivo é encontrar padrões estáveis nos dados.



### 5. Avaliação

Nesta etapa verificamos se os resultados fazem sentido.

Uma ferramenta comum é a **curva do cotovelo (elbow curve)**, usada para escolher o número de grupos.

No exemplo, **quatro grupos** descrevem bem os padrões observados.

Também se observa que:

- tumores benignos tendem a ser mais homogêneos  
- tumores malignos apresentam maior variabilidade



### 6. Implantação

Por fim, o modelo pode ser usado na prática.

Ele pode integrar um **sistema de apoio ao diagnóstico médico**.

Assim, novas amostras podem ser analisadas automaticamente.



### Resumo das etapas da CRISP-DM

1. Compreensão do problema  
2. Compreensão dos dados  
3. Preparação dos dados  
4. Modelagem  
5. Avaliação  
6. Implantação

```python
# Leitura e exploração inicial do conjunto de dados de Câncer de Mama de Wisconsin

import pandas as pd
from sklearn.datasets import load_breast_cancer

# carregando o dataset
dados = load_breast_cancer()

# criando um DataFrame
df = pd.DataFrame(dados.data, columns=dados.feature_names)

# adicionando a variável resposta (diagnóstico)
df["diagnostico"] = dados.target
```

```python

# primeiras linhas do conjunto de dados
print("Primeiras observações:")
print(df.head())

# dimensão do dataset
print("\nDimensão dos dados (linhas, colunas):")
print(df.shape)

# resumo estatístico
print("\nResumo estatístico:")
print(df.describe())

# contagem das classes
print("\nDistribuição dos diagnósticos:")
print(df["diagnostico"].value_counts())

# observação sobre as classes
print("\nLegenda:")
print("0 = tumor maligno")
print("1 = tumor benigno")
```
