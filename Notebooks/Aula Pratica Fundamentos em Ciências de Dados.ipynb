## 4. Listas, `for` e list comprehension (compreensão de listas)

### Exemplo 4 - Quadrados de uma lista

```python
numeros = [1, 2, 3, 4, 5]
quadrados = []

for x in numeros:
    quadrados.append(x**2)

print(quadrados)
```

### Exercício 6 - List comprehension

Complete a versão com list comprehension para obter o mesmo resultado.

```python
numeros = [1, 2, 3, 4, 5]

quadrados = [x**2 for x in numeros]

print(quadrados)
```

### Exemplo 5 - Filtragem com compreensão de listas

```python
numeros = [1, 2, 3, 4, 5, 6]
pares = [x for x in numeros if x % 2 == 0]
print(pares)
```

### Exercício 7 - Interpretando três códigos

Os três trechos abaixo atacam o mesmo problema: selecionar valores maiores que 10 e multiplicá-los por 3.
Execute e compare.

```python
# Código 1: for tradicional
numeros = [2, 5, 8, 11, 14, 17]
resultado1 = []

for numero in numeros:
    if numero > 10:
        resultado1.append(numero * 3)

print("resultado1 =", resultado1)
```

### Exercício 9 - Mesmo problema em listas e em NumPy

Compare agora três abordagens para o mesmo problema.

```python
# Lista + for
numeros = [2, 5, 8, 11, 14, 17]
resultado_for = []

for numero in numeros:
    if numero > 10:
        resultado_for.append(numero * 3)

print("resultado_for =", resultado_for)

# Lista + compreensão
numeros = [2, 5, 8, 11, 14, 17]
resultado_comp = [numero * 3 for numero in numeros if numero > 10]
print("resultado_comp =", resultado_comp)

# NumPy
a = np.array([2, 5, 8, 11, 14, 17])
resultado_np = a[a > 10] * 3
print("resultado_np =", resultado_np)
```

### Exercício 12 - Criação de coluna

Complete a lógica da coluna `situacao`:
- Aprovado: nota >= 7
- Recuperação: 5 <= nota < 7
- Reprovado: nota < 5

```python
df = pd.DataFrame({
    "aluno": ["Ana", "Bruno", "Carla", "Diego", "Eva"],
    "nota": [8.5, 4.0, 6.5, 9.0, 5.5],
    "frequencia": [90, 72, 80, 95, 68]
})

def classificar_nota(nota):
    if nota >= 7:
        return "Aprovado"
    elif nota >= 5:
        return "Recuperação"
    else:
        return "Reprovado"

df["situacao"] = df["nota"].apply(classificar_nota)
df
```

### Exercício 13 - Medidas resumo e contagem

```python
# Calculando a média das notas
media_notas = df["nota"].mean()

# Contando quantos alunos há em cada categoria da coluna situacao
contagem_situacao = df["situacao"].value_counts()

print("Média das notas:", media_notas)
print()
print(contagem_situacao)
```
