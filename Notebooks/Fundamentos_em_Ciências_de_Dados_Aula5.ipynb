### <b>Biblioteca Numpy

Biblioteca para computação científica, com operações com vetores e matrizes e implementação de diversas funções matemáticas

https://numpy.org/

```python
# Carregando a biblioteca

import numpy as np
```

```python
# Definindo vetores

a = np.array([[1],[2],[5]])
b = np.array([[1],[3],[3]])
a, b
```

```python
# Soma de dois vetores
a+b
```

```python
# Produto interno b^T.a
c = b.T.dot(a)
c
```

```python
# Matrizes
A = np.array([[7,5,3],[5,6,2],[2,2,8]])
B = np.array([[6,3,3],[12,6,6],[4,5,8]])
A, B
```

```python
# Multiplicação por escalar
w = 2
w*A
```

```python
# Produto de duas matrizes
A.dot(B)
```

```python
# Determinante de uma matriz

print(np.linalg.det(A), '\n', np.linalg.det(B))
```

```python
# Dimensão de uma matriz
A.shape
```

```python
# Inversa de uma matriz
A_inv = np.linalg.inv(A)
A_inv
```

```python
A_inv.dot(A)
```

```python
# Conferindo as contas
np.round(A_inv.dot(A))
```

## <b>Motivação: Dados de clientes de um banco</b>

Considere uma amostra de 500 clientes de um banco no arquivo amostra_banco_distritos.csv. Estão disponíveis as variáveis:

- Cliente: Identificador do cliente.
- Sexo: Feminino (F) ou Masculino (M)
- Idade: Idade do cliente, em anos completos.
- Empresa: Tipo da empresa em que trabalha: Pública, Privada ou Autônomo
- Salário: Salário declarado pelo cliente na abertura da conta, em reais.
- Saldo_cc: Saldo em conta corrente, em reais.
- Saldo_poupança: Saldo em poupança, em reais.
- Saldo_investimento: Saldo em investimentos, em reais.
- Devedor_cartao: Valor em atraso no cartão de crédito, em reais.
- Inadimplente: Se o cliente é considerado inadimplente atualmente (1) ou não (0), de acordo com critérios preestabelecidos.
- Distrito: Distrito de São Paulo em que mora o cliente.


<b>Faça a leitura dos dados. Desenvolva a exploração e visualização dos dados. Verifique possíveis associações entre variáveis. Visualize dados em um mapa.</b>

```python
import pandas as pd

# Leitura de dados em csv a partir de um link

url='https://raw.githubusercontent.com/cibelerusso/Datasets/main/amostra_banco_distrito.csv'

dados = pd.read_csv(url, sep=',', decimal=',')
dados.head(20)
```

```python
dados
```

```python
dados[['Empresa', 'Distrito']]
```

```python
# Listando as categorias de Empresa

pd.unique(dados['Empresa'])
```

```python
# Ordenando os dados por uma coluna

dados.sort_values(by='Empresa')
```

### Agregação (agrupamento) de dados: groupby


```python
dados.groupby(by='Distrito').mean(numeric_only=True).sort_values(by='Inadimplente', ascending=False)
```

```python
# Exclui a coluna Cliente

dados.drop('Cliente', axis=1).groupby(by='Distrito').mean(numeric_only=True)
```

## Subconjuntos lógicos de dados: loc

```python
dados.loc[dados['Distrito']=='JABAQUARA']
```

```python
dados.loc[((dados['Distrito']=='JABAQUARA') | (dados['Distrito']=='MOOCA')) & (dados['Sexo']=='M')]
```

```python
dados.loc[dados['Distrito'].isin(['JABAQUARA','MOOCA']) & (dados['Sexo']=='F')]
```

```python
dados.loc[dados['Idade']>40]
```

```python
dados.loc[~(dados['Idade']<=40)]
```

```python
dados.iloc[0,:]
```

```python
dados.iloc[0,1]
```

Exercício: Obtenha o salário médio para clientes com mais de 40 anos.

```python
# Resposta

dados.loc[~(dados['Idade']<=40)]['Salario'].mean()
```
