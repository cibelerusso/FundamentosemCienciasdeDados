## 1. Laços em Python

### 1.1 Laço `for` em Python

O laço `for` é usado quando queremos repetir um bloco de código para cada elemento de uma sequência, como listas, strings ou intervalos numéricos.



````python
# Exemplo 1: percorrendo uma lista

frutas = ["maçã", "banana", "uva"]

for fruta in frutas:
    print(fruta)
````

````python
# Exemplo 2: usando range

for i in range(5):
    print(i)
````

````python
# Exemplo 3: somando números

soma = 0

for i in range(1, 6):
    soma += i

print("Soma:", soma)
````

````python
# Percorrendo duas listas ao mesmo tempo

nomes = ["Ana", "Bruno", "Carlos"]
notas = [9, 8, 10]

for nome, nota in zip(nomes, notas):
    print(nome, nota)
````

Pythonico x Não-Pythonico

````python
# formas "pythonicas" (list comprehension)

numeros = [1, 2, 3, 4]

quadrados = [x**2 for x in numeros]

quadrados
````

````python
# forma não pythonica

numeros = [1, 2, 3, 4]

quadrados = []

for x in numeros:
    quadrados.append(x**2)

quadrados
````

## 1.2 Laço `while` em Python

O laço `while` é usado quando queremos repetir um bloco de código enquanto uma condição for verdadeira.


````python
# Exemplo 1

contador = 0

while contador < 5:
    print(contador)
    contador += 1
````

````python
# Exemplo 2: somando números
i = 1
soma = 0

while i <= 5:
    soma += i
    i += 1

print("Soma:", soma)
````

````python
## Cuidado para não gerar um loop infinito

x = 0

while x < 3:
    print(x)
    x += 1
````

## 2. Estruturas de dados em Python

Antes de trabalhar com bibliotecas científicas, é importante dominar as estruturas básicas da linguagem:

- **lista (`list`)**: coleção ordenada e mutável;
- **tupla (`tuple`)**: coleção ordenada e imutável;
- **conjunto (`set`)**: coleção sem repetição;
- **dicionário (`dict`)**: pares `chave: valor`.

Essas estruturas são úteis para organizar dados, parâmetros, rótulos e resultados.


````python
import numpy as np
import pandas as pd

print("NumPy:", np.__version__)
print("Pandas:", pd.__version__)
````

````python
# Estruturas básicas
idades = [19, 21, 20, 22, 21]
coordenadas = (3.5, 7.2)
categorias = {"A", "B", "A", "C"}
aluno = {"nome": "Ana", "idade": 21, "curso": "Ciência de Dados"}

print("Lista:", idades)
print("Tupla:", coordenadas)
print("Set:", categorias)
print("Dicionário:", aluno)

print("\nOperações comuns:")
print("Média das idades:", sum(idades) / len(idades))
print("Primeira idade:", idades[0])
print("Curso do aluno:", aluno["curso"])
````


## 3. NumPy: arrays e computação numérica

A biblioteca **NumPy** é fundamental para computação numérica em Python.  
Seu principal objeto é o **array**, que permite operações vetorizadas, rápidas e compactas.

Vantagens dos arrays NumPy:
- melhor desempenho que listas para operações numéricas;
- suporte a matrizes e tensores;
- base para várias bibliotecas de Ciência de Dados.


````python
# Criando arrays
vetor = np.array([10, 20, 30, 40, 50])
matriz = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("Vetor:", vetor)
print("Dimensão do vetor:", vetor.shape)
print("\nMatriz:\n", matriz)
print("Dimensão da matriz:", matriz.shape)

# Operações vetorizadas
print("\nVetor * 2:", vetor * 2)
print("Vetor + 5:", vetor + 5)
print("Média:", vetor.mean())
print("Desvio padrão:", vetor.std())
````

````python
# Comparando listas Python com arrays NumPy
lista_python = [1, 2, 3, 4]
array_numpy = np.array([1, 2, 3, 4])

print("Lista Python repetida por 2:", lista_python * 2)
print("Array NumPy multiplicado por 2:", array_numpy * 2)
````


## 7. Visualização de dados e tipos de gráficos

A visualização de dados ajuda a:
- explorar padrões;
- identificar outliers;
- comunicar resultados;
- apoiar a tomada de decisão.

### Tipos de gráficos mais comuns

- **gráfico de linha**: evolução no tempo;
- **gráfico de barras**: comparação entre categorias;
- **histograma**: distribuição de frequências;
- **gráfico de dispersão (scatter plot)**: relação entre duas variáveis numéricas;
- **boxplot**: dispersão, mediana e possíveis outliers.

No ecossistema Python, a base da visualização costuma ser o **Matplotlib**.  
O **Pandas** usa o Matplotlib internamente para vários gráficos.


````python

# Pequeno conjunto de dados para visualização
np.random.seed(42)

dados_vis = pd.DataFrame({
    "horas_estudo": np.random.normal(5, 1.5, 100).clip(1, 10),
    "nota": np.random.normal(7, 1.2, 100).clip(0, 10),
    "turma": np.random.choice(["A", "B", "C"], 100),
})

dados_vis["aprovado"] = np.where(dados_vis["nota"] >= 7, "Sim", "Não")
dados_vis.head()
````

````python

# 1) Gráfico de linha
media_por_indice = dados_vis["nota"].rolling(10).mean()

plt.figure(figsize=(9, 4))
plt.plot(media_por_indice)
plt.title("Gráfico de linha: média móvel da nota")
plt.xlabel("Observação")
plt.ylabel("Nota média móvel")
plt.grid(alpha=0.3)
plt.show()
````

````python

# 2) Gráfico de barras
media_por_turma = dados_vis.groupby("turma")["nota"].mean()

plt.figure(figsize=(7, 4))
plt.bar(media_por_turma.index, media_por_turma.values)
plt.title("Gráfico de barras: média de nota por turma")
plt.xlabel("Turma")
plt.ylabel("Nota média")
plt.show()
````

````python

# 3) Histograma
plt.figure(figsize=(8, 4))
plt.hist(dados_vis["nota"], bins=12)
plt.title("Histograma: distribuição das notas")
plt.xlabel("Nota")
plt.ylabel("Frequência")
plt.show()
````

````python

# 4) Gráfico de dispersão
plt.figure(figsize=(8, 4))
plt.scatter(dados_vis["horas_estudo"], dados_vis["nota"], alpha=0.7)
plt.title("Scatter plot: horas de estudo x nota")
plt.xlabel("Horas de estudo")
plt.ylabel("Nota")
plt.show()
````

````python

# 5) Boxplot
plt.figure(figsize=(8, 4))
plt.boxplot(
    [dados_vis.loc[dados_vis["turma"] == turma, "nota"] for turma in ["A", "B", "C"]],
    tick_labels=["A", "B", "C"]
)
plt.title("Boxplot: distribuição das notas por turma")
plt.xlabel("Turma")
plt.ylabel("Nota")
plt.show()
````


### Quando usar cada gráfico?

- **Linha**: séries temporais e tendências;
- **Barras**: comparação de categorias;
- **Histograma**: distribuição de uma variável;
- **Dispersão**: associação entre duas variáveis numéricas;
- **Boxplot**: dispersão, quartis e outliers.

Escolher o gráfico adequado é tão importante quanto calcular a estatística correta.



## 8. Séries temporais

Séries temporais são muito frequentes em problemas reais:
- vendas ao longo dos meses;
- sensores e IoT;
- clima;
- mercado financeiro;
- acessos a sistemas.

Frequências dos dados de séries temporais:
https://pandas.pydata.org/docs/user_guide/timeseries.html#dateoffset-objects

A análise temporal exige atenção à **ordem cronológica** dos dados.


````python

datas = pd.date_range("2025-01-01", periods=24, freq="ME")
vendas = pd.Series(
    200 + np.cumsum(np.random.randint(-15, 20, size=24)),
    index=datas
)

df_temporal = pd.DataFrame({"vendas": vendas})
df_temporal["media_movel_3"] = df_temporal["vendas"].rolling(3).mean()

display(df_temporal.head())

plt.figure(figsize=(10, 4))
plt.plot(df_temporal.index, df_temporal["vendas"], label="Vendas mensais")
plt.plot(df_temporal.index, df_temporal["media_movel_3"], label="Média móvel (3)")
plt.title("Série temporal de vendas")
plt.xlabel("Data")
plt.ylabel("Vendas")
plt.legend()
plt.grid(alpha=0.3)
plt.show()
````


## 9. Imagens como dados

Imagens podem ser tratadas como arrays NumPy:
- imagem em tons de cinza → matriz 2D;
- imagem colorida RGB → matriz 3D (`altura x largura x canais`).

Abaixo, usamos um conjunto clássico do Scikit-learn com imagens de dígitos manuscritos.


````python
from sklearn.datasets import load_digits

digits = load_digits()

print("Formato do conjunto de imagens:", digits.images.shape)
print("Formato da matriz de atributos:", digits.data.shape)
print("Classes disponíveis:", np.unique(digits.target))
````

````python
import matplotlib.pyplot as plt


plt.rcParams["figure.figsize"] = (8, 4)
pd.set_option("display.max_columns", None)
````

````python

fig, axes = plt.subplots(2, 5, figsize=(10, 4))

for ax, imagem, rotulo in zip(axes.ravel(), digits.images[:10], digits.target[:10]):
    ax.imshow(imagem, cmap="gray")
    ax.set_title(f"Rótulo: {rotulo}")
    ax.axis("off")

plt.suptitle("Exemplos de imagens do dataset digits")
plt.tight_layout()
plt.show()
````
